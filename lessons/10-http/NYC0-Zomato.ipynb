```python
import requests

def get_zomato_city(search):
    zomato_key = '8a0d66c268b134b0ee4f99a66ec7af99'
    url = 'https://developers.zomato.com/api/v2.1/cities'
    my_query_string = {'q' : search}
    my_headers = { 'user-key' : '8a0d66c268b134b0ee4f99a66ec7af99' }
    response = requests.get(url, 
        headers = my_headers, params = my_query_string)
    results = response.json()
    return results['location_suggestions'][0]


get_zomato_city('Syracuse, NY')
```

```python
def get_zomato_rest(city_id, search):
    zomato_key = '8a0d66c268b134b0ee4f99a66ec7af99'
    url = "https://developers.zomato.com/api/v2.1/search"
    my_query_string = {
        'entity_id' : city_id, 
        'entity_type' : 'city', 
        'q' : search 
    }
    my_headers = {'user-key' : zomato_key }
    response = requests.get(url, params = my_query_string, headers = my_headers)
    results = response.json()
    return results['restaurants']

get_zomato_rest(988, "pizza")
```

```python
location = input("Enter your travel destination: ")
search = input(f"What type of food are you hungry for in {location}? ")
city = get_zomato_city(location)
rests = get_zomato_rest(city['id'], search)
print("Top 10 Locations from the Zomato API: ")
for rest in rests[:10]:
    print(rest['restaurant']['name'])
    print(rest['restaurant']['location']['address'])
    print()
```
