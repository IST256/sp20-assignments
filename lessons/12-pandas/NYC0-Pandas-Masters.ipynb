```python
import pandas as pd
import requests
from pandas.io.json import json_normalize
from IPython.display import display, HTML

zomato_key = 'putYourKeyHere'
cuse_id = 988   #city_id of Syracuse, New York'  

headers = { 'user-key' : zomato_key } 
while True:
    display(HTML("<h4>Enter cuisine or type 'quit'</h4>"))
    search = input()
    if search == 'quit':
        break
        
    query_string = { 'q' : search, 'entity_id' : 988, 'entity_type' : 'city'}
    response = requests.get("https://developers.zomato.com/api/v2.1/search", params=query_string, headers=headers)
    data = response.json()
    rests = data['restaurants']
    rests_df = json_normalize(rests)
    results_df = rests_df[[ 'restaurant.name', 'restaurant.location.address', 'restaurant.user_rating.aggregate_rating','restaurant.phone_numbers']]
    results_df.columns = ['Name', 'Address', 'Rating', 'Phone']
    filter_results_df = results_df[ results_df.Rating.astype('float') > 4.0 ]
    display(HTML(f"<h3 style='text-color: blue;'>Top {search.title()} Restaurants</h3>"))
    display(filter_results_df.sort_values('Rating', ascending=False))

```
